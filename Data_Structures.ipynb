#Bubble sort

```python
grades=[10,5,6,20,15]
#اول حاجه علشان ارتب ال array بتاعي ان انا اعرف طوله
#1)length of array
length=len(grades)
print("length=  ",length) #9
print("grades before bubble sorrt to the array : " , grades)

#2) loop on element in array
for index in range(0,length-1):
  first=grades[index]
  second=grades[index +1]
  if first >second:
    grades[index]=second
    grades[index +1]=first
    print("grades in loop : " , grades)


print("grades after bubble sorrt to the array : " , grades)

#  الطريقه دي ماتنفعش لو في رقم صغير في الاخر علشان دي بيلف مره واحده بس يعني بيفضل يقارن بس الرقم اللي ماسكه اللي عليه الدور  بالرقم اللي بعده
```

#هي نفس الحاجه بس بطريقه مختلفه swap

```python
grades = [10, 50, 70, 80, 22, 50, 90, 39, 21]

# First, we determine the length of the array
# 1) Length of the array
length = len(grades)
print("Length of array: ", length)  # 9
print("Grades before Bubble Sort: ", grades)

# 2) Bubble Sort algorithm
# Loop through the array multiple times to sort it
for i in range(length - 1):  # Outer loop for passes
    for j in range(length - 1 - i):  # Inner loop for comparisons
        if grades[j] > grades[j + 1]:  # Swap if the current element is greater than the next
            grades[j], grades[j + 1] = grades[j + 1], grades[j]  # Swap the elements

print("Grades after Bubble Sort: ", grades)
```

#Linear Search

#في هذا المثال، نقوم بالبحث عن القيمة 30 في القائمة my_list. إذا تم العثور على القيمة، يتم طباعة الموقع الذي تم العثور عليه، وإذا لم يتم العثور على القيمة، يتم طباعة رسالة بأنها غير موجودة في القائمة.

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i  # يُعيد المؤشر على الموقع الذي تم العثور عليه
    return -1  # إذا لم يتم العثور على العنصر

# مثال على الاستخدام:
my_list = [10, 20, 30, 40, 50]
target_value = 30
result = linear_search(my_list, target_value)

if result != -1:
    print(f"تم العثور على {target_value} في الموقع {result}")
else:
    print(f"{target_value} غير موجود في القائمة")
```

#Binary Search

#(Binary Search)؟
اBinary Searchهو طريقة بحث فعالة تستخدم عندما تكون البيانات مرتبة. بدلاً من البحث عنصرًا عنصرًا كما في البحث الخطي، يعتمد اBinary Search على تقسيم القائمة إلى نصفين في كل خطوة:

قارن العنصر المطلوب بالعنصر الموجود في منتصف القائمة.
إذا كان العنصر المطلوب يساوي العنصر الأوسط، تنتهي العملية.

إذا كان العنصر المطلوب أصغر من العنصر الأوسط، تابع البحث في النصف الأول.

إذا كان العنصر المطلوب أكبر، تابع البحث في النصف الثاني.
كرر هذه الخطوات حتى تجد العنصر أو حتى تنتهي القائمة.

image.png

```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1  # إذا لم يتم العثور على العنصر

# مثال على الاستخدام:
sorted_list = [10, 20, 30, 40, 50]
target_value = 30
result = binary_search(sorted_list, target_value)

if result != -1:
    print(f"تم العثور على {target_value} في الموقع {result}")
else:
    print(f"{target_value} غير موجود في القائمة")
```
